### Making data frame of the questions

```python
import pandas as pd
```

```python
que_df = pd.DataFrame(test_questions)
que_df
```

```python
que_df.options
```

```python
optdf = pd.DataFrame(que_df.options.to_list(), columns=['op1', 'op2', 'op3', 'op4'])
optdf = pd.concat([pd.json_normalize(optdf[col]).add_prefix(f'{col}_') for col in optdf.columns], axis=1)
optdf.head()
```

```python
optdf.columns
```

Creatin a common df where it contains all the details of the questions

```python
detailed_df = pd.merge(que_df, optdf, left_on='id', right_on='op1_question_id', how='inner')
detailed_df.head()
```

Checking the data

```python
navals = detailed_df.isna().sum()
for col in navals.index:
    if navals[col] != 0:
        print(navals[col],"\t", col)
```

```python
detailed_df[detailed_df.type.isna()]
```
